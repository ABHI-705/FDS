```python
#declare matplotlib library
import matplotlib.pyplot as plt

#declaring the figure or the plot (y,x) or width, height)
plt.figure(figsize = (15,10))

#categorial data
countries = ['China','USA','Russia','UK','India','Spain']

#integer values for death count
totalDeaths = [52596, 37312, 59712, 45985, 76598, 56648]
plt.bar(countries, totalDeaths, width = 0.5, align = 'center', alpha = 0.5, color = ('red'))

#to define the annotations above the particular bar
i=1.0
j=2000
#define the position above the bar
for i in range(len(countries)):
    plt.annotate(totalDeaths[i],(-0.1 + i, totalDeaths[i] + j))

#add a scale on plot
plt.legend(labels = ['Total Deaths'])
plt.xticks(countries, totalDeaths)
plt.ylabel("death count")
plt.xlabel("countries")
plt.title("Total death count")

#saves the png file
plt.savefig('covid_deaths.png')

#shows the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
countries = ['China','USA','Russia','UK','India','Spain']
totalCases = [654327,548215,745932,695314,958632,756328]
totalDeaths = [52596, 37312, 59712, 45985, 76598, 56648]

#plot both the total deaths and total cases in single bar 
for i in range(len(countries)):
    plt.bar(countries[i], totalDeaths[i], bottom = totalCases[i] - totalDeaths[i], color = 'black')
    plt.bar(countries[i], totalCases[i] - totalDeaths[i], color = 'r')
    
plt.legend(labels = ['Total Deaths','Total Cases'])
plt.ylabel("total cases")
plt.xlabel("countries")
plt.title("total cases v/s death")
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
countries = ['China','USA','Russia','UK','India','Spain']
totalDeaths = [2596, 7312, 9712, 5985, 7598, 5648]
totalRecovery = [56982,45973,52894,40598,62589,55014]
totalCases = [65432,54215,74532,69314,95632,75328]

#using numpy to group 3 different data with bars
x = np.arange(len(totalDeaths))

#passing the parameters to the bar function, this is the main fun.
#using X now to align the bars side by side
plt.bar(x, totalDeaths, color = 'black', width = 0.25)
plt.bar(x + 0.25, totalRecovery, color = 'g', width = 0.25)
plt.bar(x + 0.5, totalCases, color = 'r', width = 0.25)

#create legend`
plt.legend(['Total Deaths', 'Total Recovery', 'Total Cases'])

#overiding the x axis with country names
plt.xticks([i + 0.25 for i in range(6)], countries)

#title 
plt.title('Bar plot representing the total deaths, recovery and cases')

#naming the x and y axis
plt.xlabel('Countries')
plt.ylabel('Cases')
plt.show()
```

# Pie Chart

```python
#data frame plotting
from pandas import DataFrame
import matplotlib.pyplot as plt

Data = {'Tasks':[300,500,700],
       'Task Type':['Tasks Pending','Tasks Ongoing','Tasks Completed']
       }
df = DataFrame(Data)
df.set_index('Task Type', inplace = True)

#autopct has extra % at the end as escape, as % is interpreted as formatting
#only pie charts need label to be data frame index
#'%1.3f%% it will shows the values after decimal point i.e. 3
df.plot.pie(y = 'Tasks', figsize = (10,10), autopct = '%1.1f%%', startangle = 90)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#pie chart, where the slices will be ordered and plotted counter-clockwise
labels = ['Civil','Mechanical','Electrical','Chemical']
sizes = [150,400,450,600]

fig,ax = plt.subplots()
ax.pie(sizes, labels = labels, autopct = '%1.1f%%', startangle = 90)
ax.axis('equal') #equal aspect ratio ensures the pie chart is circular
ax.set_title('Engineering Disciplines')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#if using jupyter notebook include
%matplotlib inline

#pie chart, where the slices will be ordered and plotted counter-clockwise
labels = ['Civil','Mechanical','Electrical','Chemical']
sizes = [150,400,450,250]

#Explode out the 'chemical' pie piece by offsetting it a greater amount
explode = (0.1,0.1,0.1,0.4)

fig,ax = plt.subplots()
ax.pie(sizes, explode = explode, 
       labels = labels, 
       autopct = '%1.1f%%', 
       shadow = True,
       startangle = 90)
ax.axis('equal')
ax.set_title('Engineering Disciplines')

plt.show()
```
