```python
import pandas as pd
import numpy as np
state = pd.read_csv("US_violent_crime.csv")
state.head()
```

```python
def some_fun(x):
    return x*2
state.apply(some_fun)
#update each entry by 2 and char. are repeated while int are multiplied by 2
```

```python
#lambda also works same as of some_fun
state.apply(lambda n : n*2)
```

```python
state.transform(func = lambda x : x*10)
```

```python
#groupby fun
#find the avg of Murder
#reset index use to add the index to the new o/p
mean_purchase = state.groupby('State')["Murder"].mean().rename("user_mean").reset_index()
print(mean_purchase)
```

```python
#add the column mean_purchase to the main table
merge = state.merge(mean_purchase)
print(merge)
```

```python
#check for missing/ null values
print(state.isnull().sum())
```
